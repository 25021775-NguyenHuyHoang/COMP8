```python
#Bai3:
class diem:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def khoangcach(self):
    return (self.x**2+self.y**2)**0.5
  def kiemtra(self):
    if self.x==0 and self.y==0:
      return 'la goc toa do'
    elif self.x==0:
      return 'la diem nam tren truc tung'
    elif self.y==0:
      return 'la diem nam tren truc hoanh'
x,y=map(int,input().split())
diem=diem(x,y)
print(f'M {diem.kiemtra()} va khoang cach cua M den O la {diem.khoangcach():.2f}')
```

```python
#Bai4:
class maytinh:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def add(self):
    return self.a + self.b
  def sub(self):
    return self.a - self.b
  def pro(self):
    return self.a * self.b
  def div(self):
    if self.b==0:
      return 'Khong the chia cho 0'
    else:
      return self.a / self.b
  def pow(self):
    return self.a**self.b
  def mod(self):
    return self.a%self.b
  def set_numbers(self,a,b):
    self.a=a
    self.b=b
mt=maytinh(0,0)
while True:
  h=input('ban co muon thoat?')
  if h.lower()=='yes':
    break
  a,b=map(int,input().split())
  mt.set_numbers(a,b)
  print("Chon phep tinh:")
  print("1. add")
  print("2. subtract")
  print("3. product")
  print("4. divide")
  print("5. power")
  print("6. mod")
  op = input("Nhap lua chon (1-6): ")
  if op == "1":
      print("Ket qua:", mt.add())
  elif op == "2":
      print("Ket qua:", mt.sub())
  elif op == "3":
      print("Ket qua:", mt.pro())
  elif op == "4":
      print("Ket qua:", mt.div())
  elif op == "5":
      print("Ket qua:", mt.pow())
  elif op == "6":
      print("Ket qua:", mt.mod())
  else:
      print("Lua chon khong hop le")
```

```python
#bai5:
class ShoppingCart:
    def __init__(self):
        self.danh_sach_sp = []
    def add_product(self, name, price):
        product = {"name": name, "price": price}
        self.danh_sach_sp.append(product)
        print(f"Da them '{name}' vao gio hang.")
    def remove_product(self, name):
        for product in self.danh_sach_sp:
            if product["name"].lower() == name.lower():
                self.danh_sach_sp.remove(product)
                print(f"Da xoa '{name}' khoi gio hang.")
                return
        print(f"Khong tim thay san pham '{name}'")
    def is_empty(self):
        return len(self.danh_sach_sp) == 0
    def total_price(self):
        total = sum(product["price"] for product in self.danh_sach_sp)
        return total
    def display_cart(self):
        if self.is_empty():
            print("Gio hang dang trong.")
        else:
            print("--- CAC SAN PHAM TRONG GIO ---")
            for i, sp in enumerate(self.danh_sach_sp, 1):
                print(f"{i}. {sp['name']} - Gia: {sp['price']}")
    def clear_cart(self):
        self.danh_sach_sp.clear()
        print("Da xoa toan bo gio hang.")
cart = ShoppingCart()

while True:
    thoat = input("\nCo muon thoat khoi chuong trinh khong? (yes/no): ")
    if thoat.lower() == "yes":
        print("Tam biet!")
        break

    print("\n--- MENU QUAN LY GIO HANG ---")
    print("2. Them san pham")
    print("3. Xoa san pham theo ten")
    print("4. Kiem tra gio hang trong")
    print("5. Tinh tong tien")
    print("6. Hien thi gio hang")
    print("7. Xoa sach gio hang")

    choice = input("Nhap lua chon cua ban (2-7): ")

    if choice == "2":
        ten = input("Nhap ten san pham: ")
        gia = float(input("Nhap gia san pham: "))
        cart.add_product(ten, gia)
    elif choice == "3":
        ten = input("Nhap ten san pham can xoa: ")
        cart.remove_product(ten)
    elif choice == "4":
        if cart.is_empty():
            print("Gio hang dang trong.")
        else:
            print("Gio hang dang co san pham.")
    elif choice == "5":
        print(f"Tong tien hien tai: {cart.total_price()}")
    elif choice == "6":
        cart.display_cart()
    elif choice == "7":
        cart.clear_cart()
    else:
        print("Lua chon khong hop le.")
```

```python
#Bai8:
class chuyenkhoan:
  def __init__(self,nguoigui,sotien):
    self.nguoigui=nguoigui
    self.sotien=sotien
  def withdraw(self,n):
    self.sotien-=n
    return chuyenkhoan(self.nguoigui,self.sotien)
  def deposit(self,n):
    self.sotien+=n
    return chuyenkhoan(self.nguoigui,self.sotien)
  def __str__(self):
    return f'{self.sotien:.2f}'
data=input().split()
ck=chuyenkhoan(data[0],int(data[1]))
m=int(input())
for i in range(m):
  p=input().split()
  if p[0]=='DEPOSIT':
    n=int(p[1])
    op=ck.deposit(n)
  if p[0]=='WITHDRAW':
    n=int(p[1])
    if ck.sotien>=n:
      op=ck.withdraw(n)
print(op)
```

```python
#Bai9:
class add_score:
  def __init__(self,name):
    self.name=name
    self.total_score=0
    self.count=0
  def new_score(self,score):
    self.total_score+=score
    self.count+=1
  def avg(self):
    if self.total_score==0:
      return 0
    return self.total_score / self.count
  def rank(self):
    h= self.avg()
    if h>=8:
      return 'Excellent'
    elif 6.5<=h<8:
      return 'Good'
    elif 5<=h<6.5:
      return 'Average'
    else:
      return 'Poor'
name=input()
k=int(input())
std=add_score(name)
for i in range(k):
  n=input().split()
  std.new_score(int(n[1]))
print(f'{std.name} {std.avg():.2f} {std.rank()}')
```

```python
#Bai10:
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.books = []
    def add(self, book):
        self.books.append(book)
    def count_by_author(self, name):
        count = 0
        for book in self.books:
            if book.author == name:
                count += 1
        return count
    def find_by_year(self, y):
        count = 0
        for book in self.books:
            if book.year == y:
                count += 1
        return count
def main():
    library = Library()
    try:
        n = int(input())
    except EOFError:
        return
    for _ in range(n):
        line = input().split()
        if not line:
            continue
        command = line[0]
        if command == "ADD":
            data = " ".join(line[1:]).split(';')
            title = data[0]
            author = data[1]
            year = data[2]
            new_book = Book(title, author, year)
            library.add(new_book)
        elif command == "COUNT":
            author_name = " ".join(line[1:])
            print(library.count_by_author(author_name))
        elif command == "COUNTYEAR":
            year_value = line[1]
            print(library.find_by_year(year_value))
if __name__ == "__main__":
    main()
```

```python
#Bai11:
import math
class Vector:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y, self.z - other.z)
    def cross(self, other):
        return Vector(self.y * other.z - self.z * other.y,
                      self.z * other.x - self.x * other.z,
                      self.x * other.y - self.y * other.x)
    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def absolute(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)
points = []
for _ in range(4):
    points.append(Vector(*map(float, input().split())))
A, B, C, D = points
n1 = (B - A).cross(C - B)
n2 = (C - B).cross(D - C)
cos_phi = n1.dot(n2) / (n1.absolute() * n2.absolute())
phi = math.degrees(math.acos(cos_phi))
print(f"{phi:.2f}")
```

```python
#Bai12:
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = float(realPart)
        self.imaginaryPart = float(imaginaryPart)
    def multiply(self, other):
        a = self.realPart
        b = self.imaginaryPart
        c = other.realPart
        d = other.imaginaryPart
        new_real = a * c - b * d
        new_imag = a * d + b * c
        return ComplexNumber(new_real, new_imag)
    def __str__(self):
        return f"{self.realPart:g} {self.imaginaryPart:g}"
try:
    line1 = input().split()
    if line1:
        c1 = ComplexNumber(line1[0], line1[1])
        line2 = input().split()
        if line2:
            c2 = ComplexNumber(line2[0], line2[1])
            result = c1.multiply(c2)
            print(result)
except EOFError:
    pass
```

```python
#Bai13:
import math
class Triangle:
    def __init__(self,e1,e2,e3):
        self.e1=int(e1)
        self.e2=int(e2)
        self.e3=int(e3)
    def getArea(self):
        a,b,c=self.e1,self.e2,self.e3
        if a+b>c and a+c>b and b+c>a:
            p=(a+b+c)/2
            return math.sqrt(p*(p-a)*(p-b)*(p-c))
        return -1
try:
    s=input().split()
    if len(s)==3:
        t=Triangle(s[0],s[1],s[2])
        r=t.getArea()
        if r==-1:
            print("invalid")
        else:
            if r==int(r):print(int(r))
            else:print(f"{r:.2f}")
except EOFError:
    pass
```
