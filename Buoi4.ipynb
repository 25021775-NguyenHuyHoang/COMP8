```python
#Bai7
def snt(x):
  if x<2:
    return False
  for i in range(2,int(x**0.5)+1):
    if x%i==0:
      return False
  return True
def uocnt():
  while True:
    try:
      n=int(input())
      if n>1:
        break
    except:
      print(f'hay nhap dung so nguyen duong lon hon 1!')
  h=[]
  for i in range(1,n+1):
    if n%i==0:
      if snt(i):
       h.append(i)
  return max(h)
print(uocnt())

```

```python
#Bai8
def daoso(n):
  k=0
  h=n
  while h>0:
    k=k*10 + h%10
    h=h//10
  return k
def ktdx(n):
  return n == daoso(n)
def ketqua():
  n=int(input())
  m=0
  while not ktdx(n):
    n= n + daoso(n)
    m+=1
  print(m)
  return n
print(ketqua())
```

```python
#Bai15
a=int(input('Nhap tong so con:'))
b=int(input('nhap tong so chan:'))
from sympy import symbols,Eq,solve
x,y= symbols('x y')
eq1= Eq(2*x+4*y,b)
eq2= Eq(x+y,a)
nghiem= solve((eq1,eq2),(x,y))
if nghiem[x]==int(nghiem[x]) and nghiem[y]==int(nghiem[y]):
  print(nghiem)
else:
  print('invalid')
```
