```python
#Bai9:
h=eval(input())
m=eval(input())
c={}
for i in h:
  c[i]=h[i]
for i in m:
  if i not in c:
    c[i]=m[i]
  else:
    c[i]+=m[i]
c=dict(sorted(c.items()))
print(c)
```

```python
#Bai10:
n=list(map(int,input().split()))
h=int(input())
m=[]
for i in range(len(n)):
  for j in range(i+1,len(n)):
    if n[i]+n[j]==h:
      k=tuple(sorted((n[i],n[j])))
      if k not in m:
         m.append(k)
print(m)

```
