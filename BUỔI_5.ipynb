```python
#Bai1:
def max1(a, b):
  return [a if a > b else b]
a, b = map(int, input().split())
print(*max1(a, b))
```

```python
#Bai2:
def swap(a, b) -> (int, int):
  a = a^b
  b = b^a
  a = a^b
  return a, b
x, y = map(int, input().split())
x, y = swap(x, y)
print(x, y)
```

```python
#Bai7:
def caculator(so1,so2,pheptoan):
  if pheptoan=="+":
    h= so1 + so2
  elif pheptoan=='-':
    h= so1 - so2
  elif pheptoan=='*':
    h= so1 * so2
  elif pheptoan=='/':
    if so2==0:
      return 'phep toan khong hop le'
    else:
      h= so1 / so2
  else:
    return 'phep toan khong hop le'
  return round(h ,2)
a=float(input())
b=float(input())
c=input()
print(caculator(a,b,c))
```

```python
#Bai10:
def anhxa(a, b):
    if len(a) != len(b):
        return False

    map_ab = {}
    map_ba = {}

    for i in range(len(a)):
        ca = a[i]
        cb = b[i]

        if ca in map_ab:
            if map_ab[ca] != cb:
                return False
        else:
            map_ab[ca] = cb

        if cb in map_ba:
            if map_ba[cb] != ca:
                return False
        else:
            map_ba[cb] = ca

    return True

x,y=input().split()
print(anhxa(x, y))
```
