```python
#Bai1:
def binary_seach(arr, left, right, target):
  while left <= right:
    mid=(left + right)//2
    if arr[mid]==target:
      return mid
    elif arr[mid]>target:
      right= right - 1
    else:
      left= left + 1
  return -1
n=list(map(int,input().split()))
h=int(input())
print(binary_seach(n,0,len(n)-1,h))
```

```python
#Bai3:
#bubble sort:
def bubble_sort(arr):
  for i in range(len(arr)):
    for j in range(len(arr)-1-i):
      if arr[j] > arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
  return arr
n=list(map(int,input().split()))
print(*bubble_sort(n))
```

```python
#Bai3:
#Selection sort:
def selection_sort(arr):
  for i in range(len(arr)):
    min_index=i
    for j in range(i+1,len(arr)):
      if arr[j]<arr[min_index]:
        min_index=j
    arr[i],arr[min_index]=arr[min_index],arr[i]
  return arr
arr=list(map(int,input().split()))
print(*selection_sort(arr))
```

```python
#Bai5:
def socapab(n,X):
  count=0
  for i in range(len(n)-1):
    for j in range(i+1,len(n)):
      if n[i]+n[j]==X:
        count+=1
  return count
n=list(map(int,input().split()))
X=int(input())
print(socapab(n,X))
```

```python
#Bai6:
def listnew(n):
  m=[]
  for i in n:
    if isinstance(i, list):
      m.extend(listnew(i))
    else:
      m.append(i)
  return m
n=eval(input())
print(listnew(n))
```

```python
#Bai7:
def dainhat(n):
  h=len(n)
  if h==0:
    return 0
  k=[1]*h
  for i in range(h):
    for j in range(i):
      if n[i]>n[j]:
        k[i]=max(k[i],k[j]+1)
  return max(k)
n=eval(input())
print(dainhat(n))
```

```python
#Bai8:
def lenmin(a,b):
  k=[]
  for j in b:
    dodai=[]
    for i in a:
      if i[0]<=j<=i[1]:
        m=i[1]-i[0]+1
        dodai.append(m)
    if dodai:
      k.append(min(dodai))
    else:
      k.append(-1)
  return k
a=eval(input())
b=eval(input())
print(lenmin(a,b))
```

```python
#Bai10:
import bisect
def max_subarray_mod(nums, m):
    prefix = 0
    prefixes = []
    ans = 0
    for x in nums:
        prefix = (prefix + x) % m
        idx = bisect.bisect_right(prefixes, prefix)
        if idx < len(prefixes):
            ans = max(ans, (prefix - prefixes[idx] + m) % m)
        else:
            ans = max(ans, prefix)
        bisect.insort(prefixes, prefix)
    return ans
nums=eval(input())
m=int(input())
print(max_subarray_mod(nums,m))
```

```python
#Bai12:
def sapxep(h):
  bangquyuoc={'thu hai':2,
            'thu ba':3,
            'thu tu':4,
            'thu nam':5,
            'thu sau':6,
            'thu bay':7,
            'chu nhat':1}
  kiemtra=list(h.values())[0].lower().strip()
  if kiemtra in bangquyuoc:
    ketqua=sorted(h.items(), key= lambda x: bangquyuoc.get(x[1].lower().strip()))
  else:
    ketqua=sorted(h.items(), key= lambda x: x[1].lower().strip())
  phanin=[item[0] for item in ketqua]
  return phanin
m=['A', 'B', 'C', 'D']
c={}
for i in m:
  c[i]=input()
print(*sapxep(c))
```

```python
#Bai13:
def daydainhat(n):
  h=[]
  for i in range(len(n)):
    m=[]
    m1=[]
    m1.append(n[i])
    m.append(n[i])
    for j in range(i+1,len(n)):
      if n[i]==n[j]:
        m.append(n[j])
        m1.append(n[j])
      if n[i]==n[j]-1:
        m.append(n[j])
      elif n[i]==n[j]+1:
        m1.append(n[j])
    h.append(max(len(m),len(m1)))
  return max(h)
n=list(map(int,input().split()))
print(daydainhat(n))

```

```python
#Bai14:
def so7(a,b):
  if len(b)!=a:
    b=list(map(int,input(f'vui long nhap dung {a} so nguyen:').split()))
  h=[]
  for i in range(len(b)):
    if b[i]==7:
      h.append(i)
  if len(h)==0:
    return ['Not found']
  return h[::-1]
a=int(input())
b=list(map(int,input().split()))
print(*so7(a,b))
```
